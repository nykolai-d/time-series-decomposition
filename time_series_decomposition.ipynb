# Time Series Decomposition with pandas-datareader

In this code, we download the closing prices for the following stocks:
AAPL, GOOGL, FB, AMZN, WMT, BAC, WFC, AXP, GE, GM
<br>and decompose their time series into trend, seasonality and noise, while showcasing the distribution and density of the data. <br>Also, we use data time warping to compare their paths and plot their trends, in order to visually appreciate which may have a similar behavior.

```python
import pandas as pd
import pandas_datareader as web
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a list containing the stocks
acciones = ["AAPL", "GOOGL", "FB", "AMZN", "WMT", "BAC", "WFC", "AXP", "GE", "GM"] 

# Download the stocks into 'precios'
precios = web.get_data_yahoo(acciones, 
        start = "2015-08-26",
        end = "2022-02-28")
```

```python
precios['Adj Close'].tail(1) 
```

These are the closing prices of the stocks on February 28th 2022.

```python
# Plot the behavior of the stocks' prices during the last seven years.
precios['Adj Close'].plot(figsize=(15,4))
plt.xlabel("Year")
plt.ylabel("Adjusted closing price")
plt.title("Stock prices 2016-2022")
plt.show()
```

```python
import seaborn as sns 
from statsmodels.tsa.seasonal import seasonal_decompose 
from pylab import rcParams 
```

```python
# Define a function that receives a dataframe and descomposes its timeseries into trend, seasonality and noise.
# Also, it plots the destribution and density of the data.

def ts(df):
    plt.figure(figsize=(17, 4))
    sns.distplot(df)
    plt.show()
    
    decomposition = seasonal_decompose(df, model='multiplicative', period=96) 
    rcParams['figure.figsize'] = 16, 12
    fig = decomposition.plot()
    plt.show()
```

```python
from dtw import dtw,accelerated_dtw
```

```python
# Define a function that receives two data frames and calculates the distance between their paths
# using data time warping.

def dtw(df1, df2):  
    d1 = df1.interpolate().values
    d2 = df2.interpolate().values
    d, cost_matrix, acc_cost_matrix, path = accelerated_dtw(d1,d2, dist='euclidean')
    plt.figure(figsize=(4,4))
    plt.imshow(acc_cost_matrix.T, origin='lower', cmap='gray', interpolation='nearest')
    plt.plot(path[0], path[1], 'w')
    plt.xlabel('Stock_1')
    plt.ylabel('Stock_2')
    plt.title(f'DTW Minimum Path with minimum distance: {np.round(d,2)}')
    plt.show()
```

```python
dtw(df_sp,df_amzn) # DTW between S&P500 y AMZN
```

```python
dtw(df_sp,df_googl) # DTW between S&P500 y GOOGL
```

```python
dtw(df_sp,df_fb) # DTW between S&P500 y FB
```

### Plotting and comparing the trends of the stocks

```python
def sdc(df):
    return seasonal_decompose(df, model='multiplicative', period=96)
```

```python
aapl_sd = sdc(df_aapl)
googl_sd = sdc(df_googl)
fb_sd = sdc(df_fb)
amzn_sd = sdc(df_amzn)
wmt_sd = sdc(df_wmt)
bac_sd = sdc(df_bac)
wfc_sd = sdc(df_wfc)
axp_sd = sdc(df_axp)
ge_sd = sdc(df_ge)
gm_sd = sdc(df_gm)
sp500_sd = sdc(df_sp)
```

```python
plt.figure(figsize=(15,6))
plt.plot(aapl_sd.trend,ms='2')
plt.plot(googl_sd.trend,ms='2')
plt.plot(fb_sd.trend,ms='2')
plt.plot(amzn_sd.trend,ms='2')
plt.plot(wmt_sd.trend,ms='2')
plt.plot(bac_sd.trend,ms='2')
plt.plot(wfc_sd.trend,ms='2')
plt.plot(axp_sd.trend,ms='2')
plt.plot(ge_sd.trend,ms='2')
plt.plot(gm_sd.trend,ms='2')

plt.plot(sp500_sd.trend,'g--')

plt.ylabel("Trends")
plt.title("Trend Behavior")

plt.legend(['aapl','googl','fb','amzn','wmt','bac','wfc','axp','ge','gm','sp500'])
plt.show()
```
